```python
#1Calculando el salario semanal
salario_base = 500 #se ingresa el salario base
total_ventas = float(input("Ingrese el total de las ventas")) #Se le asigna un float porque se multiplicara por un numero con decimales
comision_ventas = total_ventas * 0.10 #se multiplica el total de ventas ingresados x la comision del 10%
salario_total = salario_base + comision_ventas # se suma el salario base + la comision calculada
print("El salario es:", salario_total)
```

```python
#2Bienvenido a la fruter√≠a virtual
catalogo_frutas  = {"manzana" : "100",
                    "pera": "124",
                       "guinda" : "63",
                            "frutilla": "450"} # se ingresa diccionario
solicitud = input("Ingrese una fruta: ").lower() #se le asigna la variable lower para evitar errores
fruta = catalogo_frutas.get(solicitud, 'no existe') 
print(f"El precio de la fruta es {fruta}")
```

```python
#3Bienvenido al casino
import random as rd # se importa la libreria random y se reenombra como rd
dado_1 = rd.randint(1,6) #se asigna la variable randint para tener un numero aleatorio entre 1 y 6
dado_2 = rd.randint(1,6) #idem linea 2
resultado = dado_1 + dado_2
print(f"El resultado del dado 1 es :", dado_1)
print(f"El resultado del dado 2 es :", dado_2)
print(f"La suma de ambos es :", resultado) 
```

```python
#Calculadora de chocolate
limite = int(input("Ingrese su limite de diario de calorias:")) #se le asgina un int para que ingrese un numero entero
calorias_chocolate = int(input("Cuantas calorias tiene una porcion de chocolate:")) #idem linea 1
calculadora = limite / calorias_chocolate # se calcula el limite diario de chocolate
print(f"Tu limite diario es de chocolate es:", int(calculadora)) # se entrega el resultado
```
